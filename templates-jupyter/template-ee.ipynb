## Calculations

### Python — Buck Converter

```python
from electrical.converter.buck import design_buck

# Configure your design parameters here:
results = design_buck(
    vin=12,       # Input voltage (V)
    vout=3.3,     # Output voltage (V)
    iout=2.0,     # Output current (A)
    fsw=300e3,    # Switching frequency (Hz)
    vripple=0.01, # Output voltage ripple (V)
    il_ripple_pct=0.30,  # 30% inductor ripple
)

# Print results
for k, v in results.items():
    if isinstance(v, float):
        print(f"{k:25s}: {v:.6e}")
    else:
        print(f"{k:25s}: {v}")
```

## Interactive Analysis

Plots display inline automatically.

```python
import matplotlib.pyplot as plt
import numpy as np

# Example: inductor current waveform
t = np.linspace(0, 1/300e3, 100)
I_L = 2.0 + 0.3 * np.sin(2 * np.pi * 300e3 * t)

plt.figure(figsize=(8, 4))
plt.plot(t * 1e6, I_L)
plt.xlabel("Time (µs)")
plt.ylabel("Inductor Current (A)")
plt.title("Inductor Current Waveform")
plt.grid(True)
plt.show()
```
